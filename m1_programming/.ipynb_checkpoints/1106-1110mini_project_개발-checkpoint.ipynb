```python
# 개발할때 참고 할 솔루션
import time

# 할 일을 저장할 딕셔너리
todos = {}

def add_todo():
    title = input("할 일의 제목을 입력하세요: ")
    duration = input("예상 소요 시간을 입력하세요: ")
    todos[title] = {"duration": duration, "completed": False}
    print(f"'{title}'이(가) 목록에 추가되었습니다.")

def display_todos():
    if not todos:
        print("할 일 목록이 비어있습니다.")
    else:
        for title, info in todos.items():
            status = "완료" if info["completed"] else "미완료"
            print(f"{title} - 예상 소요 시간: {info['duration']} - 상태: {status}")

def mark_completed():
    title = input("완료한 할 일의 제목을 입력하세요: ")
    if title in todos:
        todos[title]["completed"] = True
        print(f"'{title}'이(가) 완료로 표시되었습니다.")
    else:
        print(f"'{title}'은(는) 목록에 없습니다.")

def main_menu():
    while True:
        print("\n오늘의 할 일 관리 프로그램")
        print("1: 할 일 추가하기")
        print("2: 할 일 목록보기")
        print("3: 할 일 완료 체크하기")
        print("4: 프로그램 종료하기")
        choice = input("원하는 작업의 번호를 입력하세요: ")

        if choice == "1":
            add_todo()
        elif choice == "2":
            display_todos()
        elif choice == "3":
            mark_completed()
        elif choice == "4":
            print("프로그램을 종료합니다.")
            break
        else:
            print("잘못된 입력입니다. 1, 2, 3, 또는 4를 입력하세요.")

        time.sleep(1)  # 사용자가 결과를 확인할 수 있도록 잠시 대기

if __name__ == "__main__":
    main_menu()
```
