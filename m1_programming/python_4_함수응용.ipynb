```python
# 과제15_1102 주용규님 솔루션 # prod() 함수를 사용하심
import math
def cal_factorial(x):
    y = range(1, x+1)
    return math.prod(y)

print(cal_factorial(5))
print(cal_factorial(10))
```

```python
#과제15_1102 최재빈님 솔루션

def factorial(n):
    result = 1 # 곱셈을 할땐 초기값을 0이아닌 1로 설정
    for i in range(1, n+1):
        result *= i
    return result


print(factorial(5)) 
print(factorial(10)) 
```

```python
# 도전과제17_1102 : 문자열 리스트 states를 정형화하여 아래와 같이 출력하세요.
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']
[출력]
['Alabama',
 'Georgia',
 'Georgia',
 'Georgia',
 'Florida',
 'South   Carolina',
 'West Virginia']
```

```python
# 도전과제17_1102 전유빈님 솔루션
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']

#양 옆 스페이스 먼저 지우기
n_states = [state.strip() for state in states]

result_states = []

for state in n_states:
    if state.isalpha()==False:
        space_index = state.find(' ')
        temp_state = list(state)
        new_state = ""
        for i in range(len(temp_state)):
            if temp_state[i].isalpha():
                new_state += temp_state[i]
            if i == space_index:
                new_state += temp_state[i]
        result_states.append(new_state.title())
    else:
        result_states.append(state.title())

#리스트 형식에 맞게 세로로 출력
separate = '\',\n\''
separate_result = separate.join(result_states)
print(f'[\'{separate_result}\']')
```

```python
# 도전과제17_1102 주용규님 솔루션
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']

def cleaning(text):
    cleaned_text = ''.join([x for x in text if x.isalnum() or x.isspace()])
    return cleaned_text

f_list = [cleaning(i).strip().title() for i in states]

print(f_list)
```

#### Python random 모듈 
- Python의 표준 라이브러리 중 하나로, 다양한 유형의 무작위 데이터를 생성하기 위한 함수와 도구들을 제공합니다. 여기에는 단일 숫자, 숫자 범위, 무작위 순서로 요소를 선택하거나 배열을 섞는 기능 등이 포함
1. random.random() : 0.0에서부터 1.0 사이의 실수(float)를 반환. 0은 포함되고 1은 포함되지 않음 0 <= a < 1
2. random.uniform(a, b) : 인자로 들어온 a~b 사이의 실수(float)를 반환
3. randint(a, b) : 인자로 들어온 a, b 사이의 랜덤한 정수(int)를 반환 맨끝자리 포함
4. randrange(a, b), randrange(b) :  a <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환, randrange(b)는 0 <= x < b 의 범위 내에서의 랜덤한 정수(int)를 반환
5. random.choice(seq) : 인자로 받은 리스트, 튜플, 문자열, range 에서 무작위로 하나의 원소 반환
6. random.sample(seq or set, N) : 첫번째 인자로 받은 시퀀스데이터 or set에서 N개의 랜덤하고, unique하고, 순서상관없이 인자를 뽑아서 리스트로 만들어서 반환. 예: random.sample([1, 2, 3, 4, 5], 3)은 5개의 숫자 중 3개를 임의로 선택한 리스트를 반환
7. random.shuffle(seq) : 데이터의 순서를 무작위로 랜덤하게 바꾸어 주는 함수

```python
import random

# 인풋은 추가할 메뉴(str), 메뉴리스트(set)
def add_menu_items(item,menu):
    menu.add(item)
    return menu

# 인풋은 메뉴리스트(set), 추천수(숫자)
def get_recommendation(menu,recommend_num):
    rec_menu = random.sample(sorted(menu),recommend_num)
    return rec_menu

    
def main():
    menu = {'맥도날드', '쉑쉑버거', '편의점', '분식집', '닭곰탕', '짬뽕'}
    # 메뉴 추가
    while 1:
        menu_input = input(f"주변 음식점 이름을 적으세요. 없으면 'stop'을 입력하세요: ")
        if menu_input == 'stop':
            break
        else:
            menu = add_menu_items(menu_input,menu)
            print(f"메뉴가 성공적으로 추가되었습니다.\n<메뉴목록>\n{menu}")

    # 메뉴 추천
    end_chr = 1 # 1==메뉴 추천 루프를 실행, 0==메뉴 추천 루프를 종료
    while end_chr == 1:
        menu_num = 0
        while 1:
            menu_num_str = input("추천받고 싶은 메뉴의 갯수를 말하세요 (숫자만): ")
            if menu_num_str.isdigit() == False:
                print("잘못 입력하셨습니다. 숫자만 입력해주세요")
                continue
            if int(menu_num_str) > len(menu):
                print("잘못 입력하셨습니다. {menu.len()} 이하의 숫자만 입력해주세요")
                continue
            menu_num = int(menu_num_str)
            break
        while 1:
            rec_menu = get_recommendation(menu,menu_num)
            print(f'추천 메뉴는 {rec_menu} 입니다.')
            diff_menu_yn = input("\n다른 추천을 원하시나요? (y/n) ")
            if diff_menu_yn == "y":
                for i in rec_menu:
                    menu.remove(i)   # 중복 추천을 피할 수 있게 추천하는 메뉴 삭제
                if len(menu) ==0:
                    print("더이상 추천할 수 있는 메뉴가 없습니다.")
                    end_chr = 0
                    break
                menu_num = random.randint(1,len(menu))
                print(f"추천 메뉴의 수를 {menu_num}으로 설정합니다.")
            elif diff_menu_yn == "n":
                print("메뉴 추천을 종료합니다.")
                end_chr = 0
                break
            else:
                print("죄송합니다. 잘못입력하셨으므로 메뉴 추천을 종료합니다.")
                end_chr = 0
                break
                
main()
```

```python
# 사용자 함수 활용

import random

def add_menu_items(menu):
  print(f'기본으로 제공하는 메뉴:\n{menu}\n')
  while True:
    user_input = input("주변 음식점 이름을 적으세요. 없으면 'stop'을 입력: ").strip()
    if user_input.lower() == 'stop':
      break
    else:
      menu.add(user_input)
  return list(menu)
def get_recommendation(menu, num_items):
  try:
    picks = random.sample(menu, num_items)
    print(f'추천 메뉴는 {picks} 입니다.\n')
  except ValueError:
    print(f'숫자가 너무 큽니다. 메뉴는 {len(menu)}개밖에 없습니다.')

def main():
  menu = {"짬뽕", "분식집", "맥도날드", "쉑쉑버거", "편의점", "닭곰탕"}
  menu = add_menu_items(menu)
  while True:
    try:
      num_recommendations = int(input('추천받고 싶은 메뉴의 개수를 입력하세요: '))
      get_recommendation(menu,num_recommendations)
      break
    except ValueError:
      print(f'숫자를 적어주세요')
  choice = input('다른 추천을 원하시나요? (y/n): ').strip().lower()
  if choice == 'y':
    recmd_num = random.randrange(1,len(menu)+1)   
    print(f'제가 추천하는 메뉴수는 다음과 같이 {recmd_num}개 입니다')
    get_recommendation(menu,recmd_num)
  elif choice == 'n':
    print('식사 맛있게 하세요~')
  else:
    print('잘못된 입력입니다. 프로그램 종료합니다.')

if __name__ == "__main__":
  main()
```

```python
# 과제3_1103 서영우님 솔루션
import time
from IPython.display import clear_output

def start():
    phase_list = {'수정':0, '찾기':1}
    txt = input('글을 입력해주세요.')
    while 1:
        time.sleep(1)
        clear_output()
        phase = input("문자 수정을 원하시면 '수정'을 입력하시고, 문자 개수를 찾기 및 삭제를 원하시면 '찾기'를 입력해주세요>수정 글을 입력해주세요")
        if phase in phase_list:
            txt = fix_txt(txt) if phase_list[phase] == 0 else find_txt(txt)
            return txt
        else:
            print('(수정/찾기)중에서 입력 바랍니다.')
        
def fix_txt(txt):
    time.sleep(1)
    clear_output()
    while 1:
        which_txt = input('글에서 수정하고 싶은 문자를 입력해주세요')
        if which_txt in txt:
            while 1:
                time.sleep(1)
                clear_output()
                change_txt = input('문자를 어떻게 수정하시겠습니까?')
                if change_txt.isalnum() == True:
                    txt = txt.replace(which_txt ,change_txt)
                    time.sleep(1)
                    clear_output()
                    print(f'수정된 문자열은 : {txt}')
                    return txt
                if change_txt.isalnum() == False:
                    clear_output()
                    print('입력하신 문자가 올바르지 않습니다. 재시도 해주십시요!')
                else:
                    clear_output()
                    print('입력하신 문자가 없습니다. 재시도 해주십시요!')
        else:
            clear_output()
            print('해당 문자가 텍스트에 없습니다. 재시도 해주십시요!')

def find_txt(txt):
    count = 0
    while 1:
        time.sleep(1)
        clear_output()
        find_str = input('글에서 찾고싶은 문자를 입력해주세요.')
        if find_str in txt:
            print(f'해당 글에 존재하는 입력하신 문자의 개수는 {txt.count(find_str)}개 입니다.')
            while 1:
                delete_yn = input('해당 글에서 입력하신 문자를 모두 삭제하시겠습니까?(Y/N)')
                if delete_yn.lower() == 'y' :
                    time.sleep(1)
                    clear_output()
                    txt = txt.replace(find_str, '')
                    print(f'수정된 문자열은 : {txt}')
                    return txt
                elif delete_yn.lower() == 'n':
                    time.sleep(1)
                    clear_output()
                    print(f'수정된 문자열은 : {txt}')
                    return txt
                else:
                    clear_output()
                    print('y/n 중에서 입력바랍니다.')
        else:
            clear_output()
            print('해당 문자가 텍스트에 없습니다. 재시도 해주십시요!')


start()
```

#### re 모듈

- 파이썬에서 정규 표현식(Regular Expressions)을 지원하기 위한 표준 라이브러리.
- 정규 표현식은 텍스트에서 문자열 검색, 매칭, 치환 및 파싱을 할 때 사용하는 강력한 문자열 처리 도구

- re 모듈을 사용하여 수행할 수 있는 기본 작업

- re.match(): 문자열의 시작부터 정규식과 일치하는 패턴을 찾습니다. 일치하는 객체를 반환하거나, 일치하는 것이 없으면 None을 반환

- re.search(): 문자열 전체에 걸쳐 정규식과 일치하는 패턴을 찾는다. match와 달리 문자열의 중간부터 일치하는 패턴을 찾을 수도 있고 일치하는 객체를 반환하거나, 일치하는 것이 없으면 None을 반환

- re.findall(): 문자열 내에서 정규식과 일치하는 모든 부분 문자열을 리스트로 반환

- re.finditer(): findall()과 유사하지만, 일치하는 모든 결과를 이터레이터(iterator)로 반환하여 메모리를 절약

- re.sub(): 문자열 내에서 정규식과 일치하는 부분을 다른 문자열로 치환 # replace() 와 유사

- re.compile(): 정규식 패턴을 컴파일하여 나중에 재사용할 수 있는 정규식 객체를 생성. 이것은 패턴을 여러 번 사용할 때 성능을 최적화하는 데 도움

- 정규 표현식 패턴은 특수 문자와 시퀀스를 포함할 수 있으며, 복잡한 문자열 처리를 위한 강력한 문법을 제공.
- 예를 들어, \d는 숫자를, \w는 단어 문자를, .는 거의 모든 문자를 나타낸다(개행 문자는 제외). 대괄호([])는 문자 집합을, 
- 괄호(())는 그룹을 정의한다. 별표(*), 더하기(+), 물음표(?), 중괄호({}) 등은 일치하는 횟수를 정의.

```python
import re
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']

def clean_strings(strings):
    result = []
    for value in strings :
        value = value .strip()
        value = re.sub('[!#?]', '',value)
        value = value.title()
        result.append(value)
    return result
print(clean_strings(states)) # 프린트는 일자로 출력해주는 기능이 있음..
clean_strings(states)
```

#### 날짜 함수

- 날짜와 시간을 처리하기 위한 주요 라이브러리는 datetime입니다. 이 라이브러리는 날짜, 시간, 시간대와 관련된 다양한 클래스와 함수를 제공하며, 날짜와 시간 연산에 매우 유용
- datetime.date: 날짜(연, 월, 일)를 표현하는 클래스
    - today(): 오늘의 날짜를 반환
- datetime.time: 시간(시, 분, 초, 마이크로초)을 표현하는 클래스
- datetime.datetime: 날짜와 시간을 모두 포함하는 클래스
    - now(): 현재의 날짜와 시간을 반환
- datetime.timedelta: 두 날짜/시간 간의 차이를 표현하는 클래스
- datetime.tzinfo: 시간대 정보를 위한 기본 클래스

- datetime 모듈의 날짜와 시간 객체들을 문자열 형식으로 포맷팅하거나 문자열로부터 날짜와 시간 객체를 생성하기 위해 사용하는 주요 메서드는 strftime() -> string for time.. 시간을 스트링으로..  그리고 strptime() -> string을 파씽(그형태로바꾼다) time

- strftime() (String Format Time)
    - 날짜와 시간 객체를 문자열로 변환하는 데 사용
    - 사용자가 지정한 형식 문자열을 기반으로 날짜와 시간 정보를 문자열로 표현
- strptime() (String Parse Time)
    - 문자열을 날짜와 시간 객체로 변환하는 데 사용
    - 주어진 형식 문자열을 기반으로 문자열에서 날짜와 시간 정보를 파싱 

```python
import datetime

# 현재 날짜와 시간 가져오기
now = datetime.datetime.now()
print(now) # 2023-11-03 15:15:29.504707

# 오늘의 날짜만 가져오기
today = datetime.date.today()
print(today) # 2023-11-03

# 특정 날짜와 시간을 직접 지정
d= datetime.datetime(2023, 11, 2, 14 ,29, 45)
print(d) # 2023-11-02 14:29:45

# 날짜 간의 차이 계산
delta = datetime.timedelta(days=7) # 7일의 기간
new_date = today + delta
print(new_date) # 2023-11-10 오늘로부터 7일 후의 날짜

# 두 날짜 간의 차이 계산
date1 = datetime.date(2001, 5, 2)
date2 = datetime.date(2028, 11, 3)
diff = date2 - date1
print(diff.days) # 364
```

- 시스템 시간에 접근하거나 간단한 타이머 기능을 구현
- 함수와 같이 프로그램을 일정시간동안 중지시키는 데 사용
- 시간을 연도,월,일,시,분,초 등의 구성 요소로 변환하고 다룰 수 있는 struct_time 객체를 제공
- time 모듈은 날짜와 시간을 처리하는 다양한 프로그래밍 요구에 대응할 수 있는 기본적인 기능을 제공하지만, 더 복잡한 날짜 시간 처리나 시간대 변환등은 datetime 모듈 또는 라이브러리를 사용

```python
# strftime 함수는 시간에 관계된 것을 세밀하게 표현할 수 있는 여러 가지 포맷 코드 제공
# time.localtime() 함수는 초로 표현된 시간을 구조화된 시간 튜플 형태로 치환
import time
d = time.strftime('%d', time.localtime(time.time()))
m = time.strftime('%m', time.localtime(time.time()))
y = time.strftime('%Y', time.localtime(time.time()))
a = time.strftime('%A', time.localtime(time.time())) # 요일
print(d)
print(m)
print(y)
print(a)
```

```python
# str -> datetime -> timestamp
s = "2021-07-20 11:13:58"
timestamp = time.mktime(datetime.strptime(s,"%Y-%m-%d %H:%M:%S").timetuple())
# mktime 에 dateime을 집어넣어줘서 1차적 작업을 해주고 timetuple을 적용해줌
print(timestamp,type(timestamp))
```

```python
# datetime -> timestamp
d = datetime.today() # 만약 주어진게 datetime 이라면...
print(d,type(d))
timestamp = time.mktime(d.timetuple())
print(timestamp, type(timestamp))
```

#### 과제9_1103 : 파이썬의 random 모듈을 사용하여 간단한 구매 데이터 세트를 생성하세요. 이 데이터 세트에는 다음과 같은 정보가 포함:

- 구매 ID # 중복되면 안됨 # 무조건 랜덤
- 구매자 ID # 구매자가 다시 또 구매 구매 하면 중복될 수 있음 대신(구매자 중복 가능) 구매 ID는 중복 X # 구매자가 입력하는 id가 될수있음 # 예를 들면 전화번호..
- 구매한 제품의 ID
- 구매한 제품의 수량
- 구매 날짜와 시간
- purchase_id: 1000에서 9999 사이의 랜덤 정수
- buyer_id: 100에서 999 사이의 랜덤 정수
- product_id: 10에서 99 사이의 랜덤 정수
- quantity: 1에서 10 사이의 랜덤 정수
- purchase_date: 최근 365일 이내의 랜덤 날짜 및 시간

```python
# 과제9_1103 주용규님 솔루션
import random
import time
from datetime import datetime, timedelta

# 데이터 세트의 갯수 입력
data_num = 100

def random_date():
    # 365일 Unix 시간은 31536000 초
    ran_timestamp = random.uniform(0, 31536000)
    ran_time = time.time() - ran_timestamp
    return (datetime.fromtimestamp(ran_time)).strftime('%Y-%m-%d %H:%M:%S')

purchase_id = list(random.randint(1000,9999) for i in range(data_num))
buyer_id = list(random.randint(100,999) for i in range(data_num))
product_id = list(random.randint(10,99) for i in range(data_num))
quantity = list(random.randint(1,10) for i in range(data_num))
purchase_date = list(random_date() for i in range(data_num))

data_set = list(zip(purchase_id, buyer_id, product_id, quantity, purchase_date))

# 데이터 세트 중에서 10개만 출력
for i, data in enumerate(data_set):
    if i == 10: break
    print(data)
```
